```python
import pandas as pd
from skimpy import skim
pd.set_option("display.max_columns", None)
# from ydata_profiling import ProfileReport
```

```python
df = pd.read_csv("../data/acled/2025.csv")
df.info()
```

```python
df.sort_values(["fatalities"], ascending=False).head(30)
```

```python
df[df["fatalities"] > 0]["country"].value_counts().to_frame("n").assign(
    **{
        "%": lambda x: (x["n"] / x["n"].sum() * 100).round(2),
        "cum%": lambda x: x["%"].cumsum(),
    }
).head(10)
```

```python
df[df["fatalities"] > 0]["event_type"].value_counts().to_frame("n").assign(
    **{
        "%": lambda x: (x["n"] / x["n"].sum() * 100).round(2),
        "cum%": lambda x: x["%"].cumsum(),
    }
)
```

```python
df[df["fatalities"] > 0]["sub_event_type"].value_counts().to_frame("n").assign(
    **{
        "%": lambda x: (x["n"] / x["n"].sum() * 100).round(2),
        "cum%": lambda x: x["%"].cumsum(),
    }
)
```
