```python
df = pd.read_csv("../data/conflicts.csv")
df.sample(n=5)
```

```python
def get_news_results(query):
    """Function to call the NewsAPI and return total results."""
    try:
        return newsapi.get_everything(q=query)['totalResults']
    except Exception as e:
        print(f"Could not retrieve news for '{query}': {e}")
        return 0
```

```python
get_news_results('sudan civil war')
```

```python
df['n_newsapi'] = df['search_query'].progress_apply(get_news_results)
df
```

```python
df.sort_values(by = ['n_newsapi'], ascending = False)
```
